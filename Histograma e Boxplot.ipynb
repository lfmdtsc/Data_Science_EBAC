```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook
```

```python
sinasc_ro_2019 = pd.read_csv(
    'sinasc_ro_2019.csv')
sinasc_ro_2019.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.boxplot(y="IDADEMAE", data=sinasc_ro_2019[sinasc_ro_2019['IDADEMAE']>0])
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(y="PESO", data=sinasc_ro_2019[sinasc_ro_2019['PESO']>0])
```

####  3) Faça um histograma da idade da mãe

```python
sns.displot(data = sinasc_ro_2019[sinasc_ro_2019['IDADEMAE']>0]
            ,alpha = .25
            ,x = 'IDADEMAE'
            ,bins = 20
            ,element = 'step'
            ,kde = True
           )
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(data = sinasc_ro_2019[sinasc_ro_2019['PESO']>0]
            ,alpha = .25
            ,x = 'PESO'
            ,bins = 20
            ,element = 'step'
            ,kde = True
           )
```
