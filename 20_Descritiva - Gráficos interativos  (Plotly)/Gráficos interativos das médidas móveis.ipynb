#### 1. Carregue os dados da tarefa anterior

***Importação dos Pacotes:***

````python
from pandas_datareader import data as web
from pandas_datareader import data as pdr

import plotly.express as px
from plotly import graph_objects

from matplotlib import pyplot
import yfinance as yf
import pandas as pd
import numpy as np
````

````python
data = pd.read_csv('./AAPL-TSLA-GOOG.csv', header=0) #carregando o dataframe do exercício anterior.

data = data.rename(columns = {'Unnamed: 0':'Date'}) #renomeando a coluna date.

datas = data[('Date')] #realizando uma cópia externa da coluna date.

new_index = pd.MultiIndex.from_tuples([('Adj Close', 'AAPL'), ('Adj Close', 'GOOG'), ('Adj Close', 'TSLA'),  
('Close', 'AAPL'),      
('Close', 'GOOG'),     
('Close', 'TSLA'),      
('High', 'AAPL'),       
('High', 'GOOG'),     
('High', 'TSLA'),      
('Low', 'AAPL'),     
('Low', 'GOOG'),     
('Low', 'TSLA'),     
('Open', 'AAPL'),      
('Open', 'GOOG'),      
('Open', 'TSLA'),      
('Volume', 'AAPL'),    
('Volume', 'GOOG'),       
('Volume', 'TSLA')]) #montando um conjunto de tuplas com o MultiIndex.

data = (data.reset_index()
           .drop(columns=['Date'])
           .drop(columns=['index'])) #resetando o index e excluindo as colunas date e index.

data.columns = new_index #modifiando os nomes das colunas com as tuplas.

date = pd.DataFrame(datas, columns=['Date']) #transformando a series date em df para realizar a inclusão no df data.

data['Date'] = date #reinserindo a coluna date.

data = (data.drop(labels=[0,1], axis=0)
            .set_index('Date')) #excluindo as duas primeiras linhas do dataframe e colocando a coluna date como index do df.

data
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu

2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
data_ac = data['Adj Close'].copy() #isolando as colunas relacionadas ao Adj Close.
data_ac
````

````python
data_ac_mean = data_ac.rolling(90).mean() #calculando a média móvel com a função rolling para 90 dias.
data_ac_mean
````

````python
#calculando o desvio padrão com a função rolling para 90 dias. FIQUEI COM DÚVIDA SE CALCULAVA O STD PARA O DF (data_ac) 
# OU PARA O DF (data_ac_mean), UMA VEZ QUE A FUNÇÃO ROLLING JÁ PERMITE O CALCULO DIRETO DO DESVIO PADRÃO.?????????
data_ac_std = data_ac_mean.rolling(90).std() 
data_ac_std
````

````python
data_ac_mean_vertical = data_ac_mean.stack().reset_index()
data_ac_mean_vertical.columns = ['data', 'ativo', 'média_móvel']

fig = px.line(data_ac_mean_vertical, x="data", y="média_móvel", color='ativo', template='plotly_white')
fig.show()
````

````python
data_ac_std_vertical = data_ac_std.stack().reset_index()
data_ac_std_vertical.columns = ['data', 'ativo', 'desvio_padrão']

fig = px.line(data_ac_std_vertical, x="data", y="desvio_padrão", color='ativo', template='plotly_white')
fig.show()
````

**Analisando os gráficos acima com o gráfico do exercício anterior, nota-se uma correlação entre eles, principalmente nos períodos em que a amplitude (apresentada no exercício 01) aumentava e reduzia.**

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
lista_indicadores = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume'] #criando uma lista com os indicadores
lista_indicadores
````

````python
ativo = ['TSLA']*len(lista_indicadores) #criando uma lista com o nome do ativo que será multiplicado pelo numero de colunas da lista de indicadores.
ativo
````

````python
colunas = list(zip(lista_indicadores, ativo)) #o comando list zip cria a tupla, indicando as colunas que serão "zipadas".
colunas
````

````python
tsla = data.loc['2023-04-28':,colunas] #selecionando o trecho do dataframe desejado
tsla.columns = lista_indicadores #indicando as colunas desse novo dataframe pela lista de indicadores.
tsla
````

````python
graph = {
    'x': tsla.index,
    'open': tsla.Open,
    'close': tsla.Close,
    'high': tsla.High,
    'low': tsla.Low,
    'type': 'candlestick',
    'name': 'TSLA',
    'showlegend': True
}
````

````python
from plotly import graph_objects
from plotly.graph_objs import Layout

layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data = [graph],
    layout_title="Índice NASDAQ",
    layout=layout
)
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
