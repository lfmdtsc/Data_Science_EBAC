####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(20, 15))
fig, axes = plt.subplots(2, 3)


ax1 = sns.countplot(ax = axes[0, 0], x='APGAR1', data=sinasc_ro_2019)
ax2 = sns.countplot(ax = axes[0, 1], x='APGAR5', data=sinasc_ro_2019)
ax3 = sns.countplot(ax = axes[1, 0], x='SEXO', data=sinasc_ro_2019)
ax4 = sns.countplot(ax = axes[1, 1], x='GESTACAO', data=sinasc_ro_2019)
ax5 = sns.countplot(ax = axes[1, 2], x='ESTCIVMAE', data=sinasc_ro_2019)

plt.subplots_adjust(wspace=0.5, hspace=0.5)

ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, horizontalalignment='right', fontweight='light')
ax4.set_xticklabels(ax4.get_xticklabels(), rotation=45, horizontalalignment='right', fontweight='light')

```
