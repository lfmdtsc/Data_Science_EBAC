````python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
previsao_de_renda.info()
````

````python
previsao_de_renda['data_ref'] = pd.to_datetime(previsao_de_renda['data_ref'], format ='%Y-%m-%d')
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
ax = sns.countplot(x='data_ref',  hue = 'educacao', data=previsao_de_renda)

tick_labs = previsao_de_renda['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

# tick_labs.tolist()

ticks = ax.set_xticks(list(range(previsao_de_renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref',  hue = 'tipo_renda', data=previsao_de_renda)

tick_labs = previsao_de_renda['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

# tick_labs.tolist()

ticks = ax.set_xticks(list(range(previsao_de_renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref',  hue = 'sexo', data=previsao_de_renda)

tick_labs = previsao_de_renda['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

# tick_labs.tolist()

ticks = ax.set_xticks(list(range(previsao_de_renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
tab = pd.crosstab(previsao_de_renda['data_ref'],previsao_de_renda['sexo'])
tab.div(tab.sum(axis = 1), axis = 0)
````

````python
tab_freq = pd.crosstab(previsao_de_renda['data_ref'],previsao_de_renda['sexo'])
tab = tab_freq.div(tab.sum(axis = 1), axis = 0)

ax = tab.plot.bar(stacked = True)
````

````python
tab_1 = pd.crosstab(previsao_de_renda['data_ref'],previsao_de_renda['posse_de_veiculo'])
tab_1.div(tab.sum(axis = 1), axis = 0)
````

````python
tab_1_freq = pd.crosstab(previsao_de_renda['data_ref'],previsao_de_renda['posse_de_veiculo'])
tab_1 = tab_1_freq.div(tab.sum(axis = 1), axis = 0)

ax = tab_1.plot.bar(stacked = True)
````

***Os gráficos acima mencionados, não apresetaram nenhuma variação significativa ao longo do tempo com relação as variáveis verificadas.

````python
ax = sns.pointplot(x="data_ref", y="posse_de_veiculo", hue = 'sexo',
                   data=previsao_de_renda, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

tick_labs = previsao_de_renda['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(previsao_de_renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
````

````python
ax = sns.pointplot(x="data_ref", y="posse_de_imovel", hue = 'sexo',
                   data=previsao_de_renda, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

tick_labs = previsao_de_renda['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(previsao_de_renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
````

***Ao analisar a variável posse de imóvel no tempo, pode-se perceber que houve uma variação da média no tempo, no sexo masculino, entre 06-2015 e 11-2015 e também um decaimento no ano de 2016 (entre janeiro e março).

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'sexo',
                   data=previsao_de_renda, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

tick_labs = previsao_de_renda['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(previsao_de_renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
````
