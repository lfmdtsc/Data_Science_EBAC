### 3. Melhores 3 variáveis

Rode uma árvore com profundidade máxima igual a 4 para prever a atividade humana com todas as variáveis.
Observe a importância das variáveis e considere as 3 variáveis com maior importância para os próximos passos.

````python
clf = DecisionTreeClassifier(random_state=2360873, max_depth = 4).fit(X_train_1, Y_train_1)
````

````python
clf.feature_importances_
````

````python
feat_importances = pd.Series(clf.feature_importances_, index=X_train_1.columns)
feat_importances.nlargest(3).plot(kind='barh')
````

### 4. Construa uma árvore com as 3 melhores variáveis

Utilizando as três variáveis encontradas acima, construa uma árvore de decisão. Encontre o melhor ```ccp_alpha``` utilizando a base de validação, conforme a estrutura que vimos em aula.

````python
X_train_1 = X_train_1[['tGravityAcc-energy()-X', 'fBodyAcc-mean()-X', 'angle(Y,gravityMean)']]
````

````python
X_valid = X_valid[['tGravityAcc-energy()-X', 'fBodyAcc-mean()-X', 'angle(Y,gravityMean)']]
````

````python
caminho = DecisionTreeClassifier(random_state=2360873).cost_complexity_pruning_path(X_train_1, Y_train_1)
ccp_alphas, impurities = caminho.ccp_alphas, caminho.impurities
````

````python
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=2360873, ccp_alpha=ccp_alpha).fit(X_train_1, Y_train_1)
    clfs.append(clf)
````

````python
train_scores = [clf.score(X_train_1, Y_train_1) for clf in clfs]
valid_scores = [clf.score(X_valid, Y_valid) for clf in clfs]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("Acurácia")
ax.set_title("Acurácia x alpha do conjunto de dados de treino e validação")
ax.plot(ccp_alphas, train_scores, marker='o', label="validação",
        drawstyle="steps-post")
ax.plot(ccp_alphas, valid_scores, marker='o', label="treino",
        drawstyle="steps-post")
ax.legend()
plt.show()
````

````python
ind_melhor_arvore = len(valid_scores) - valid_scores[::-1].index(max(valid_scores)) - 1
melhor_arvore = clfs[ind_melhor_arvore]

print(f'Acurácia da melhor árvore na base de validação: {valid_scores[ind_melhor_arvore]*100:.2f}')

melhor_arvore
````

### 5. Avaliação do modelo

Avalie a árvore encontrada no item anterior na base de testes.

````python
X_test = X_test[['tGravityAcc-energy()-X', 'fBodyAcc-mean()-X', 'angle(Y,gravityMean)']]
````

````python
acuracia_teste = melhor_arvore.score(X_test, Y_test)

print(f'Acurácia: {acuracia_teste*100:.2f}')
plot_confusion_matrix(melhor_arvore, X_test, Y_test);
````

A acurácia da base de teste caiu significativamente em relação a base de validação, 87,21 contra 69,94. O exercício com o maior número de acerto foi o 6, o 4 e o 5 tiveram algumas amostras que se confundiram entre si durante a avaliação.
